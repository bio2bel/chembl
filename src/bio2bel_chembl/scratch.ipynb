```python
url_dir_chembl = "ftp://ftp.ebi.ac.uk/pub/databases/chembl/ChEMBLdb/latest/chembl_uniprot_mapping.txt"
```

```python
import pandas as pd
from pybel.utils import ensure_quotes
```

```python
df =pd.read_csv(url_dir_chembl, skiprows=1, sep="\t", names=['uniprot', 'chembl', 'name', 'type'])
```

```python
df.head()
```

```python
df['uniprot']
```

```python
for _, gfam, gene in df[['name', 'type']].itertuples():
     gfam_clean = ensure_quotes(gfam.strip())
     gene_clean = ensure_quotes(gene.strip())
     print('g(HGNC:{}) isA g(GFAM:{})'.format(gene_clean, gfam_clean))
```
